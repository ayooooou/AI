```python
import pandas as pd
import csv

#讀取資料
train_df=pd.read_csv('train.csv')
test_df=pd.read_csv('test.csv')
```

```python
#刪除不要的資料
train_df = train_df.drop(['Name','Cabin','Ticket'], axis=1)
test_df = test_df.drop(['Name','Cabin','Ticket'], axis=1)
```

```python
#把資料變得可計算
train_df['Sex'] = train_df['Sex'].replace({'male':  0, 'female': 1})
test_df['Sex'] = test_df['Sex'].replace({'male' : 0, 'female': 1})
train_df['Embarked'] = train_df['Embarked'].replace({'S':1, 'C': 2, 'Q': 3})
test_df['Embarked'] = test_df['Embarked'].replace({'S': 1, 'C':2, 'Q': 3})
```

```python
#把空白資料填進平均值
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].mean())
train_df['Embarked'] = train_df['Embarked'].fillna(0)
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].mean())
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].mean())
```

```python
train_df.head(10)
```

```python
#訓練的(題目:x 答案:y)
x_train=train_df.drop(['Survived'],axis=1)
y_train=train_df['Survived']
```

```python
x_train.head(10)
```

```python

y_train.head(10)
```

```python
#把x_train,y_train(100%)集分成 X_train,Y_train(訓練集80%)和 X_test Y_test(測試集20%)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x_train,y_train,test_size=0.2,shuffle=True)
```

```python
#定義model 為隨機樹模型
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=100,max_depth=40)
```

```python
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import mean_squared_error
test_preds=model.predict(X_test)
score=mean_squared_error(test_preds,Y_test)
print('mse score: ',score)
```
